<h2 style="color:Black" align="center">The survival prediction in The Titanic sink</h2>
<h3 align="center">using descision tree</h3>

```python
import pandas as pd
```

**reading the dataset**

```python
df = pd.read_csv("titanic.csv")
df.head()
```

**Finding 'NaN' values in 'Age' and filling with the mean of all values**

```python
import math
median_age = math.floor(df.Age.mean())
df.Age = df.Age.fillna(median_age)
df
```

**Droping 'PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked' as of no use**

```python
inputs = df.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
inputs.head()
```

```python
target = df['Survived']
target.head()
```

**lebelling male as 1 and female as 0 using 'LabelEncoder'**

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
```

```python
inputs['Sex_n'] = le_Sex.fit_transform(inputs['Sex'])
```

```python
inputs.head()
```

```python
inputs_n = inputs.drop('Sex',axis='columns')
inputs_n
```

**import tree to use 'DecisionTreeClassifier()' method**

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```
